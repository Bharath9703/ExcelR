```python
import numpy as np
import scipy.stats as stats

# Sample data
data = np.array([1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85,
                 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29])

# a. Using sample standard deviation
n = len(data)
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # Sample standard deviation

# t-score for 99% confidence interval with n-1 degrees of freedom
t_critical = stats.t.ppf(0.995, df=n-1)

# Margin of Error
margin_of_error_sample = t_critical * (sample_std / np.sqrt(n))

# Confidence Interval
ci_sample = (sample_mean - margin_of_error_sample, sample_mean + margin_of_error_sample)

print(f"99% Confidence Interval using Sample Standard Deviation: {ci_sample}")

# b. Using known population standard deviation
population_std = 0.2  # Known population standard deviation

# z-score for 99% confidence interval
z_critical = stats.norm.ppf(0.995)

# Margin of Error
margin_of_error_population = z_critical * (population_std / np.sqrt(n))

# Confidence Interval
ci_population = (sample_mean - margin_of_error_population, sample_mean + margin_of_error_population)

print(f"99% Confidence Interval using Known Population Standard Deviation: {ci_population}")
```
