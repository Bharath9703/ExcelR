```python
# Install the required package before running


import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

# Load dataset
file_path = 'Online retail.xlsx'
retail_data = pd.read_excel(file_path)

# Pre-process the dataset by splitting the product lists into individual transactions
transactions = retail_data[retail_data.columns[0]].str.split(',')
transactions_cleaned = transactions.dropna()

# Convert the transactions into a one-hot encoded DataFrame
te = TransactionEncoder()
te_ary = te.fit(transactions_cleaned).transform(transactions_cleaned)
df_trans = pd.DataFrame(te_ary, columns=te.columns_)

# Apply the Apriori algorithm with a minimum support threshold of 0.01 (1%)
frequent_itemsets = apriori(df_trans, min_support=0.01, use_colnames=True)

# Generate the association rules with minimum confidence and lift thresholds
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.3)

# Sort rules by lift to find the strongest rules
rules_sorted = rules.sort_values(by="lift", ascending=False)

# Display the top 5 rules
print(rules_sorted.head())
```

1. Lift in Association Rules
    Definition: Lift measures how much more often two items appear together than if they were statistically independent. It's calculated as the ratio of the observed joint probability to the expected joint probability if the two events were independent.




2. Support and Confidence
    Support: The proportion of transactions that contain a certain item or item set. High support indicates common occurrences in the dataset.

    Confidence: The likelihood that if a transaction contains item A, it also contains item B. Confidence helps assess the reliability of the rule.
